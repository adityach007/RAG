```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
documents = SimpleDirectoryReader('/content/').load_data()
```

```python
documents
```

```python
from llama_index.core import PromptTemplate

system_prompt = """

<|SYSTEM|># You are a Q&A assistant. Your goal is to answer the questions as accurately possible based on the instructions and context provided.

"""

# This is a type of default prompt style in llamaindex
query_wrapper_prompt = PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
import torch

llm = HuggingFaceLLM(
    context_window = 4096,
    max_new_tokens = 256,
    generate_kwargs = {"temperature": 0.7, "do_sample": False},
    system_prompt = system_prompt,
    query_wrapper_prompt = query_wrapper_prompt,
    tokenizer_name = "mistralai/Mistral-7B-Instruct-v0.1",
    model_name = "mistralai/Mistral-7B-Instruct-v0.1",
    device_map = "auto",
    stopping_ids = [50278, 50279, 50277, 1, 0],
    tokenizer_kwargs = {"max_length": 4096},
    model_kwargs={"torch_dtype": torch.float16}
)
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
embed_model = HuggingFaceEmbedding(model_name = "sentence-transformers/all-mpnet-base-v2")
```

```python
from llama_index.core import VectorStoreIndex , Settings
from llama_index.core.node_parser import SentenceSplitter

Settings.llm = llm
Settings.embed_model = embed_model
Settings.node_parser = SentenceSplitter(chunk_size=512, chunk_overlap=20)
```

```python
index = VectorStoreIndex.from_documents(documents, embed_model = embed_model)
```

```python
query_engine = index.as_query_engine()
```

```python
query_engine.query("what is RAG?")
```
